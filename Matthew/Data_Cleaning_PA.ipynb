```python
#importing 
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
```

```python
data_csv_path = "Resources/Physical Activity.csv"
initial_data = pd.read_csv(data_csv_path)
print(len(initial_data))
del initial_data['Unnamed: 10']
del initial_data['*Methodological changes in the 2011 BRFSS may affect comparability between the 2011 estimates and estimates from earlier years']
initial_data.head()
```

```python
#Dropping District of Colombia
initial_data["State"].unique()
```

```python
US_PA_data = initial_data.iloc[:1]
US_PA_data = US_PA_data.drop(['County'], axis = 1)
US_PA_data.head()
```

```python
state_county_PA = initial_data.drop(labels = 0,
                                   axis = 0,
                                   inplace = False)
state_county_PA = state_county_PA.drop([327,328])
state_county_PA.head()
```

```python
#Exporting US dataframe and county's Dataframe to CSV
state_county_PA.to_csv("Data_csv/state_county_PA.csv")
US_PA_data.to_csv("Data_csv/US_PA.csv")
```

```python
#Creating a dataframe for the Obesity by State
state_PA_df = state_county_PA.loc[state_county_PA['County'].isna()]
state_PA_df = state_PA_df.reset_index()
del state_PA_df['index']
del state_PA_df['County']
state_PA_df
```

```python
state_PA_df.to_csv("Data_csv/state_PA_avg.csv")
```

```python
#Seperating the Data by regions
Northeast_PA = state_PA_df[state_PA_df["State"].isin(['Maine', 'Massachusetts', 'Rhode Island','Connecticut',
                                                                 'New Hampshire', 'Vermont', 'New York', 
                                                                 'Pennsylvania', 'New Jersey', 'Delaware', 'Maryland'])]

Southeast_PA = state_PA_df[state_PA_df["State"].isin(['West Virginia', 'Virginia', 'Kentucky', 'Tennessee',
                                                                   'North Carolina', 'South Carolina', 'Georgia', 'Alabama',
                                                                   'Mississippi', 'Arkansas', 'Louisiana', 'Florida'])]
Midwest_PA = state_PA_df[state_PA_df["State"].isin(['Ohio', 'Indiana', 'Michigan', 'Illinois',
                                                                 'Missouri', 'Wisconsin', 'Minnesota', 'Iowa',
                                                                 'Kansas', 'Nebraska', 'South Dakota', 'North Dakota'])]
Southwest_PA = state_PA_df[state_PA_df["State"].isin(['Texas', 'Oklahoma', 'New Mexico', 'Arizona'])]

West_PA = state_PA_df[state_PA_df["State"].isin(['Colorado', 'Wyoming', 'Montana', 'Idaho',
                                                              'Washington', 'Oregon', 'Utah', 'Nevada',
                                                              'California', 'Alaska', 'Hawaii'])]
```

```python
#Exporting regional Data
Northeast_PA.to_csv("Data_csv/Northeast_PA.csv")
Southeast_PA.to_csv("Data_csv/Southeast_PA.csv")
Midwest_PA.to_csv("Data_csv/Midwest_PA.csv")
Southwest_PA.to_csv("Data_csv/Southwest_PA.csv")
West_PA.to_csv("Data_csv/West_PA.csv")
```
