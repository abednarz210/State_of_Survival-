```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator


# DATA 
Life_Expectancy_by_Gender = ["Male life expectancy, 2000", "Male life expectancy, 2010", "Female life expectancy, 2000", "Female life expectancy, 2010"]
Years = [74.1, 76.1, 79.2, 80.8] 

x_axis = np.arange(0, len(Life_Expectancy_by_Gender))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)
    



plt.title("Life Expectancy Change 2000 to 2010")
plt.xlabel("Gender")
plt.xticks(tick_locations, Life_Expectancy_by_Gender)
plt.xticks(rotation=90)

plt.xlim(-0.5, len(Life_Expectancy_by_Gender)-.5)
plt.ylabel("Years")
plt.ylim(0, max(Years)+10)

plt.bar(x_axis, Years, facecolor="green", alpha=0.5, align="center")
plt.figure(figsize=(8,5))
plt.show()
```
