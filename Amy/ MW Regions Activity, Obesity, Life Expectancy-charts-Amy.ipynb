```python
 # Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
 # Load in csv
# Take in all of our Regions data and read it into pandas
midwest_life = "Data_csv/Midwest_life_expect.csv"
midwest_or = "Data_csv/Midwest_obes.csv"
midwest_act = "Data_csv/Midwest_PA.csv"

midwest_life_df = pd.read_csv(midwest_life)
midwest_or_df = pd.read_csv(midwest_or)
midwest_act_df = pd.read_csv(midwest_act)


midwest_act_df.head()
```

```python
# Merge the first two datasets on "State" so that no data is lost 
combined_mw_regions_df = pd.merge(midwest_act_df, midwest_or_df,
                                 how='left', on='State')
combined_mw_regions_df.head()
```

```python
# Delete the duplicate 'Unnamed x and y' columns and rename the first one back to 'Country Code'

del combined_mw_regions_df["Unnamed: 0_y"]
combined_mw_regions_df = combined_mw_regions_df.rename(columns={"Unnamed: 0_x":"State Code"})
combined_mw_regions_df
```

```python
# Create a group based on the values in the 'state' column
mw_male = combined_mw_regions_df.groupby(['State','Male sufficient physical activity prevalence, 2001 (%)', 'Male obesity prevalence, 2001 (%)'])

count_male = mw_male['State'].count()
count_male
```

```python
# Create a group based on the values in the 'state' column
mw_female = combined_mw_regions_df.groupby(['State','Female sufficient physical activity prevalence, 2001 (%)', 'Female obesity prevalence, 2001 (%)'])

count_female = mw_female['State'].count()
count_female
```
